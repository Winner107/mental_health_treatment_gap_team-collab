## Load The Dataset And Make A Copy

```python
raw_data = pd.read_csv('Downloads/mental-illnesses-prevalence.csv')
raw_data
```

```python
mh_data
```

## Framework For Measuring Trends Across Multiple Conditions.

```python
mh_data_long = pd.melt(mh_data, 
                       id_vars=['Country', 'Code', 'Year'],
                       var_name='Disorders',
                       value_name='Prevalence_rate'
    
)

mh_data_long
```

```python
mh_data_long['Disorders'] = mh_data_long['Disorders'].str.extract(r'^(.*? disorders)')[0] 
mh_data_long
```

## Calculate Year-Over-Year Prevalence Changes

```python
mh_data_long.sort_values(['Country', 'Code', 'Year'])
```

```python
mh_data_long['YoY_change'] = mh_data_long.groupby(['Country', 'Disorders'])['Prevalence_rate'].diff()
```
