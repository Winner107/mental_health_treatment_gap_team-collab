## Import Necessary Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load The Dataset And Make A Copy

```python
raw_data = pd.read_csv('Downloads/mental-illnesses-prevalence.csv')
raw_data
```

```python
mh_data = raw_data.copy()
mh_data
```

```python
pd.isnull(mh_data).count()
```

```python
mh_data.rename(columns = {
    'Entity': 'Country'}, inplace=True)
```

```python
mh_data
```

## Framework For Measuring Trends Across Multiple Conditions.

```python
mh_data_long = pd.melt(mh_data, 
                       id_vars=['Country', 'Code', 'Year'],
                       var_name='Disorders',
                       value_name='Prevalence_rate'
    
)

mh_data_long
```

```python
mh_data_long['Disorders'] = mh_data_long['Disorders'].str.extract(r'^(.*? disorders)')[0] 
mh_data_long
```

## Calculate Year-Over-Year Prevalence Changes

```python
mh_data_long.sort_values(['Country', 'Code', 'Year'])
```

```python
mh_data_long['YoY_change'] = mh_data_long.groupby(['Country', 'Disorders'])['Prevalence_rate'].diff()
```

```python
mh_data_long.pop('yoy_change')
```

```python
mh_data_long
```

```python
mh_data_long=mh_data_long.fillna(0)
mh_data_long
```

```python
mh_data_long.tail()
```

```python
mh_data_long
```

## Creating Visual Display For Year-over-Year Prevalence Changes

```python
average_yoy = mh_data_long.groupby(['Disorders', 'Year'])['YoY_change'].mean().reset_index()
```

```python
average_yoy
```

```python
mh_data_long
```

```python
conditions = average_yoy['Disorders'].unique()
conditions
```

```python
plt.figure(figsize=(12,6))

for c in conditions:
    subset = average_yoy[average_yoy['Disorders'] == c]
    plt.plot(subset['Year'], subset['YoY_change'], label = c)

plt.title('Average YoY Disorder Prevalence vs Year')
plt.legend(loc = 'lower left')
plt.xlabel('Year')
plt.ylabel('Average YoY Change')
plt.show()

plt.savefig('Average_YoY_Disorder_Prevalence_vs_Year.png')
```

## Cohort Analysis Tracking

```python
base_year = mh_data_long['Year'].min()
base_year
```

```python
mh_data_base = mh_data_long[mh_data_long['Year'] == base_year].copy()
mh_data_base
```

```python
mh_data_base['Quartile'] = mh_data_base.groupby('Disorders')['Prevalence_rate'].transform(lambda x: (pd.qcut(x, 4, labels=[1,2,3,4])))
mh_data_base
```

```python
mh_data_long = mh_data_long.merge(
    mh_data_base[['Country','Disorders', 'Quartile']],
    on = ['Country','Disorders'],
    how = 'left'
)
mh_data_long
```

```python
mh_data_long['Quartile'] = mh_data_long['Quartile'].astype(float)
mh_data_long
```

```python
average_quartile = mh_data_long.groupby(['Disorders','Quartile', 'Year'])['Prevalence_rate'].mean().reset_index()
average_quartile
```

```python
disorder_to_plot = 'Schizophrenia disorders'
```

```python
plot_data = average_quartile[average_quartile['Disorders'] == disorder_to_plot]
plot_data
```

```python
plt.figure(figsize=(12,6))

for q in sorted(plot_data['Quartile'].unique()):
    schiz_set = plot_data[plot_data['Quartile'] == q]
    plt.plot(schiz_set['Year'], schiz_set['Prevalence_rate'], marker = "s", label = f'Quartile {int(q)}')

plt.title('Prevalence Over Time by Initial Quartile')
plt.xlabel('Year')
plt.legend()
plt.ylabel('Prevalence Rate')
plt.show()
```
